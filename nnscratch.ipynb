```python
def relu(x):
    return(np.maximum(x,0))
def sigmoid(x):
    return(1/(1+np.exp(-x)))
def softmax(x):
    return(np.exp(x)/sum(np.exp(x)))
```

```python
def forwardpass(x,W1,b1,W2,b2):
    Z1 = W1.dot(x)+ b1
    A1 = relu(Z1)
    Z2 = W2.dot(A1)+b2
    A2 = softmax(Z2)
    return(Z1,A1,Z2,A2)
```

```python
def one_hot(Y):
    one_hot_Y = np.zeros((Y.size, Y.max() + 1))
    one_hot_Y[np.arange(Y.size), Y] = 1
    one_hot_Y = one_hot_Y.T
    return one_hot_Y
```

```python
def backpass(y,x,Z1,A1,Z2,A2,w1,b1,w2,b2):
    m =  60000
    dZ2 = A2 - one_hot(y)
    dW2 = 1/m* dZ2.dot(A1.T)
    db2 = 1/m * np.sum(dZ2)
    drelu= Z1>0
    dZ1 = w2.T.dot(dZ2)*drelu
    dW1 = 1/m* dZ1.dot(x.T)
    db1 = 1/m* np.sum(dZ1)
    return(dW1,db1,dW2,db2)
```

```python
def get_predictions(A2):
    return np.argmax(A2, 0)

def get_accuracy(predictions, Y):
    print(predictions)
    print(Y)
    return np.sum(predictions == Y) / Y.size
```

```python
def grad_desc(X,Y,alpha,num_iterations):
    W1 = np.random.rand(10, 784) - 0.5
    b1 = np.random.rand(10, 1) - 0.5
    W2 = np.random.rand(10, 10) - 0.5
    b2 = np.random.rand(10, 1) - 0.5
    for i in range(num_iterations):
        Z1,A1,Z2,A2 = forwardpass(X,W1,b1,W2,b2)
        dW1,db1,dW2,db2 = backpass(Y,X,Z1,A1,Z2,A2,W1,b1,W2,b2)
        print(f'this is change in W2 :{dW2.shape}')
        #print(dW1,db1,dW2,db2)
        W1 -= alpha*dW1
        b1 -= alpha*db1
        W2 -= alpha*dW2
        b2 -= alpha*db2
        if i % 10 == 0:
                print("Iteration: ", i)
                predictions = get_predictions(A2)
                print(get_accuracy(predictions,Y))
    return(W1,b1,W2,b2)
```

```python
data = pd.read_csv('./Desktop/singlecell/check/archive/mnist_train.csv')
data = np.array(data)
m, n = data.shape
np.random.shuffle(data) # shuffle before splitting into dev and training sets

data_dev = data[0:1000].T
Y_dev = data_dev[0]
X_dev = data_dev[1:n]
X_dev = X_dev / 255.

data_train = data[1000:m].T
Y_train = data_train[0]
X_train = data_train[1:n]
X_train = X_train / 255.
_,m_train = X_train.shape
```

```python
W1,b1,W2,b2 = grad_desc(X_train,Y_train, 0.1, 500)
```

```python
def make_predictions(X,W1,b1,W2,b2):
    Z1,A1,Z2,A2 = forwardpass(X,W1,b1,W2,b2)
    predictions = get_predictions(A2)
    return(predictions)
test_predictions = make_predictions(X_dev,W1,b1,W2,b2)
print(get_accuracy(test_predictions,Y_dev))
```
